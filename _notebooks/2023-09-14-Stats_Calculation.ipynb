```python
import requests
from bs4 import BeautifulSoup

#region Getting money stats

# url being parsed
url = "https://www.boxofficemojo.com/year/world/2022/"
# list where money is going to be stored
test = []
# removing dollar symbol for formatting
char = '$'
# makes a request for the website 
moviePage = requests.get(url)
# parses for all the content from web page
soup = BeautifulSoup(moviePage.content, 'html.parser')
# money is stored in specific tag, finds that specific tag
elements = soup.find_all('td', class_='a-text-right mojo-field-type-money')

# searches through beautiful soup output and removes any undesired tags
for element in elements:
    test.append(element.get_text(strip=True))

# gets rid of dollar symbol in list
for idx, ele in enumerate(test):
    test[idx] = ele.replace(char, '')

x = 0
test = test[:20]

for i in test:
#Formatting to convert list to integers
    i = i.replace(',','')
    test[x] = int(i)
    x+=1
n_num = test
#endregion

#calculating minimum and maximum
n = len(n_num)
n_num.sort()

minimum = ('{:,}'.format(n_num[0]))
maximum = ('{:,}'.format(n_num[-1]))

print("The least amount of money movies made in box office is: " + str(minimum)+ "$")
print("The most amount of money movies made in box office is: " + str(maximum)+ "$")

#calculating median
n = len(n_num)
n_num.sort()
 
if n % 2 == 0:
    median1 = n_num[n//2]
    median2 = n_num[n//2 - 1]
    median = ('{:,}'.format((median1 + median2)/2))
else:
    median = ('{:,}'.format(n_num[n//2]))
print("The Median of the amount of money movies made in box office is: " + str(median)+ "$")

#calculating mean
n = len(n_num)
 
get_sum = sum(n_num)
mean = ('{:,}'.format(get_sum / n))
 
print("The Average amount of money movies made in box office is: " + str(mean)+ "$")

#calculating standard deviation
n = len(n_num)

get_sum = sum(n_num)
mean = get_sum / n

variance = sum([((x - mean) ** 2) for x in n_num]) / n
sd = ('{:,}'.format(variance ** 0.5))

print("The Standard Deviation of the amount of money movies made in box office is: " + str(sd) + "$")
```
